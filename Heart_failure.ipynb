```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
from mlxtend.plotting import plot_decision_regions
import missingno as msno
from pandas.plotting import scatter_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
df.head()
```

```python
df[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']] = df[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']].replace(0, np.NaN)
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
df['anaemia'].fillna(df['anaemia'].mean(), inplace=True)
df['creatinine_phosphokinase'].fillna(df['creatinine_phosphokinase'].median(), inplace=True)
df['diabetes'].fillna(df['diabetes'].median(), inplace=True)
df['ejection_fraction'].fillna(df['ejection_fraction'].median(), inplace=True)
df['high_blood_pressure'].fillna(df['high_blood_pressure'].median(), inplace=True)
df['platelets'].fillna(df['platelets'].median(), inplace=True)
df['serum_creatinine'].fillna(df['serum_creatinine'].median(), inplace=True)
df['serum_sodium'].fillna(df['serum_sodium'].median(), inplace=True)
df['sex'].fillna(df['sex'].median(), inplace=True)
df['smoking'].fillna(df['smoking'].median(), inplace=True)
df['time'].fillna(df['time'].median(), inplace=True)
df.isnull().sum()
```

```python
X = df.drop(columns='DEATH_EVENT')
y = df['DEATH_EVENT']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=20)
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
accuracy 
```
